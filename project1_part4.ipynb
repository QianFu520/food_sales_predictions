```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/content/sales_predictions.csv")
df.head()
```

```python
#change inconsistent categories of data
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("LF", "Low Fat")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("reg", "Regular")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("low fat", "Low Fat")
df["Item_Fat_Content"].value_counts()
```

# Plots that show outlets play crucial roles in increasing sales

```python
plt.figure(figsize = (50,30))
sns.barplot(data=df, x='Item_Type', y = "Item_Outlet_Sales", hue = "Outlet_Type", ci=None)
plt.xticks(rotation=30, fontsize = 30)
plt.yticks(fontsize = 30)
plt.legend(loc=(1.02,0), fontsize = 30)
plt.xlabel("Item_Type", fontsize = 50)
plt.ylabel("Item_Outlet_Sales", fontsize = 50)
plt.title("Different Item_Type Sales in Different Outlets", fontsize = 60, fontweight = "bold")
```

```python
plt.figure(figsize = (50,30))
sns.lineplot(data=df, x='Item_Type', y = "Item_Outlet_Sales", hue = "Outlet_Type", ci=None)
plt.xticks(rotation=30, fontsize = 30)
plt.yticks(fontsize = 30)
plt.legend(loc=(1.02,0), fontsize = 30)
plt.xlabel("Item_Type", fontsize = 50)
plt.ylabel("Item_Outlet_Sales", fontsize = 50)
plt.title("Different Item_Type Sales in Different Outlets", fontsize = 60, fontweight = "bold")
```

Both above plots show that the sales of different type of item in different types of outlet. I can see that the supermarket type 3 has the most sales of all the items, followed by supermarket type 1, supermartket type2, and grocery. This means that the outlets play crucial roles in increasing sales.

# Plots show that properties of products play crucial roles in increasing sales.

The obove plot shows the two types of product properties have correlation with sales. so the product properties does play roles on increasing or decreasing sales.
